```python
# Ask user for name
name = input("Enter your name: ")

# Ask user for grade out of 35
grade_35 = float(input("Enter your grade out of 35: "))

# Convert grade out of 35 to grade out of 10
grade_10 = (grade_35 / 35) * 10

# Print the result
print(f"{name}, your grade over 10 is: {grade_10:.2f}")
```
