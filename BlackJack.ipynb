```python
from random import choice

# Card values like your lists
cards = [2,3,4,5,6,7,8,9,10,10,10,10,11] * 4

def hand_total(hand):
    total = sum(hand)
    aces = hand.count(11)
    while total > 21 and aces:
        total -= 10
        aces -= 1
    return total

while True:
    # Deal
    player = [choice(cards), choice(cards)]
    dealer = [choice(cards), choice(cards)]
    
    # Player turn - like your input loops
    while True:
        print(f"Your hand: {player} (total: {hand_total(player)})")
        print(f"Dealer shows: {dealer[0]}")
        
        if hand_total(player) > 21:
            print("You bust!")
            break
        action = input("Hit or stand? (h/s): ").lower()
        if action == 'h':
            player.append(choice(cards))
        else:
            break
    
    if hand_total(player) <= 21:
        # Dealer turn
        while hand_total(dealer) < 17:
            dealer.append(choice(cards))
        print(f"Dealer hand: {dealer} (total: {hand_total(dealer)})")
        
        player_total = hand_total(player)
        dealer_total = hand_total(dealer)
        
        if dealer_total > 21 or player_total > dealer_total:
            print("You win!")
        elif player_total < dealer_total:
            print("Dealer wins!")
        else:
            print("Push!")
    
    if input("\nAgain? (y/n): ").lower() != 'y':
        break
```
